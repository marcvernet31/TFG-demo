```python
punctuations = "!”#$%&'()*+,-./:;?@[\]^_`{|}~’"
batch += ["l", "d", "\\n"]

def preprocess(text):
    # remove possible nan
    if type(text) == float: text = "-"
        
    # remove \n
    text = text.strip()
    
    # remove puntuations
    for p in punctuations:
        try:
            text = text.replace(p, ' ')
        except:
            print(text)
        
    # lower text
    text = text.lower()

    # remove stopwords
    text_split = text.split(' ')
    text = ' '.join([x for x in text_split if x not in batch])
    
    return text

def gen_pairwise_similarity(text):
    vect = TfidfVectorizer(min_df=1)
    tfidf = vect.fit_transform(text)  

    pairwise_similarity = tfidf * tfidf.T 
    return(pairwise_similarity)

def similar(data, pairwise_similarity, indx, n):
    title_1 = pairwise_similarity[indx].toarray()
    title_1 = list(title_1[0])
    
    title_1_sorted = sorted(title_1, reverse=True)
    sorted_ind = [title_1.index(i) for i in title_1_sorted]
    
    print(data[indx])
    print("---")
    for i in range(n):
        if(sorted_ind[i] != indx):
            print(data[sorted_ind[i]])
            
def preprocess_list(text_list):
    return [(lambda x: preprocess(x))(x) for x in text_list]
```

```python
titles = list(catalog_unified["title"])
descriptions = list(catalog_unified["description"])

titles_clean = preprocess_list(titles)
descriptions_clean = preprocess_list(descriptions)

pairwise_similarity_titles = gen_pairwise_similarity(titles_clean)
pairwise_similarity_descriptions = gen_pairwise_similarity(descriptions_clean)
```
